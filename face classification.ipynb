```python
from deepface import DeepFace
import cv2

cap = cv2.VideoCapture(1)

while True:
    state, frame = cap.read()
    if state != True:
        break
        
    res = DeepFace.find(frame, db_path="./face", enforce_detection=False, model_name='VGG-Face')
    
    if res and len(res) > 0 and len(res[0]['identity']) > 0:
        name = res[0]['identity'][0].split('/')[2].split('.')[0]
        xmin = int(res[0]['source_x'][0])
        ymin = int(res[0]['source_y'][0])
        
        w = res[0]['source_w'][0]
        h = res[0]['source_y'][0]
        
        xmax = int(xmin + w)
        ymax = int(ymin + h)
              
        cv2.rectangle(frame, (xmin, ymin), (xmax +20, ymax),(0, 255, 0), 2)
        
        cv2.rectangle(frame, (xmin, ymin -40), (xmax + 20, ymin), (108, 245, 0), -1)
        cv2.putText(frame, name, (xmin, ymin), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
    
    cv2.imshow('attendance', frame)
    key = cv2.waitKey(2)
    
    if key == 27:
        break

cv2.destroyAllWindows()
cap.release()
```
